```python
from nltk.corpus import PlaintextCorpusReader
corpus_root = r"D:/NLP"
corpora = PlaintextCorpusReader(corpus_root,['《三国演义》.txt'],encoding='utf-8')
corpora.fileids() #查看语料库中的文本名称
myfile = corpora.words('《三国演义》.txt')

text = ''.join(myfile) #由于从语料库读取的文本无法直接用于jieba分词处理，故需要通过代码corpora.words(corpora.fileids())将其分割成链表并定义为变量myfile，接着又以代码'’.join(myfiles)将其连接成字符串并定义为变量text,此时的字符串方可用于jieba分词处理。
import jieba.analyse #加载jieba的analyse模块，为了后续提取关键词之用
seg_text = jieba.analyse.extract_tags(text,topK=1000) #extract_tags()是提取关键词函数,text为待提取的语料文本，topK=1000为输出1000个关键词
print(" ".join(seg_text)) #把上述1000个关键词连接成字符串并打印在操作界面上

bbb = " ".join(seg_text) #后续将代码print"".join(seg_text)修改为bbb="".join(seg_text),即把1000个关键词连接成字符串并定义为变量bbb。
import codecs #用于转换自然语言的编码，目的是启用open()函数。
f = codecs.open('D:/NLP/zzzz2.txt','w',encoding='utf-8')#将编码转换成utf-8并在D盘NLP文件夹下创建文件zzzz2.txt，这一过程被定义为变量f；w表示将内容写入该文件
f.write(bbb)#把变量bbb所包含的内容信息写入变量f；write（）函数用于写入文件。
f.close() #关闭变量f所表示的文件。
```
