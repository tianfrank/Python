```python
import pkuseg
import re
with open('../data/student_corpora/all_together/Japanese_students_abstract.txt','r',encoding='utf-8') as f:
    text1 = f.read()
    text1 = re.sub('[a-zA-Z\d\,\;\'，○——？！……①②③‘’“”；、：\-（）()/\n\t]',' ',text1)  # 去掉文本中的字母、数字、标点、换行符，不缩进表示在文件外操作
seg = pkuseg.pkuseg()           # 以默认配置加载模型,post=True表示的是标注词性
text1 = seg.cut(text1)  # 进行分词[(),(词，v)]
```

```python
from collections import Counter
word_dict1 = dict(Counter(text1))
```

```python
#返回值为字典的写法
morethan10={}    #{'词'：[10,20,30,40]，‘语’：[11,12,13,12]}
for word,freq in word_dict1.items():
# if freq>10 and word_dict2.get(word,0)>10 and word_dict3.get(word,0)>10 and word_dict4.get(word,0)>10
    if freq>10:
        if word_dict2.get(word,0)>10:  #key是键，0为默认返回值
            if word_dict3.get(word,0)>10:
                if word_dict4.get(word,0)>10:
                    morethan10[word]=[freq,word_dict2[word],word_dict3[word],word_dict4[word]]
with open('morethan10.txt','w',encoding='utf-8') as f:
    for word,freq in morethan10.items():
        #f.write(word+','+str(freq)+'\n') 
        f.write(word+':'+','.join([str(i) for i in freq])+'\n')  # '词':10,20,30,40 
```
