```python
import pandas as pd
data_1 = pd.read_excel('词频.xlsx')
data_2 = pd.read_excel('新HSK一至六级词汇表(2011年).xls')
data_1.head()
```

```python
data_2.head()
```

```python
names = ['词汇']
for i in range(1,6):
    name = '词汇.%s'%i
    names.append(name)
names
```

```python
words_2 = []
for name in names:
    words = list(data_2[name].map(str))
    words_2 += words
words_2 = [word.strip() for word in words_2]
words_2 = set(words_2)
words_2.remove('nan')
len(words_2)
```

```python
data_1 = data_1.dropna(axis=0,how='any')
words_1 = set(list(data_1['word'])[:5000])
len(words_1)
```

```python
words_set_1 = words_1 - words_2    # 只属于汉语教学的词语
words_set_2 = words_2 - words_1    # 只属于六级词汇的词语
common_words = words_1 & words_2   # 共有的词语
```

```python
print('只属于汉语教学的词语的个数：{}'.format(len(words_set_1)))
print('只属于六级词汇的词语的个数：{}'.format(len(words_set_2)))
print('共有的词语的个数：{}'.format(len(common_words)))
```

```python
words_set_2
```

```python
with open('只属于汉语教学的词语.txt','w',encoding='utf-8') as f:
    for word in words_set_1:
        f.write(word + '、')
        
with open('只属于六级词汇的词语.txt','w',encoding='utf-8') as f:
    for word in words_set_2:
        f.write(word + '、')
        
with open('共有的词语.txt','w',encoding='utf-8') as f:
    for word in common_words:
        f.write(word + '、')
```
