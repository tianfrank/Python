```python
import sklearn
from sklearn import datasets #语料库的包

trainData = datasets.load_files('D:/NLP/python与语料库/myCorpus/trainingData',encoding='utf-8')
testData = datasets.load_files('D:/NLP/python与语料库/myCorpus/testData',encoding='utf-8')
#print(trainData)
#print(testData)
```

```python
X_train = trainData.data
y_train = trainData.target
X_test = testData.data

print(y_train)
#print(X_train)
#print(X_test)
```

```python
from sklearn.feature_extraction.text import CountVectorizer  #导入特征抽取模块的子模块
textVector = CountVectorizer() #调用向量化的函数
print(textVectorizer )

#textVector =  CountVectorizer(stop_words=['the'],max_features=10)

X_train = textVector.fit_transform(X_train) #将文档转换为数字
X_test = textVector.transform(X_test)
```

```python
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import train_test_split

iris = datasets.load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=0)

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```

```python
from sklearn import datasets
from sklearn import tree
from sklearn.model_selection import train_test_split
from sklearn import svm

iris = datasets.load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.4, random_state=0)

clf = svm.SVC(gamma='scale')
clf = clf.fit(X_train, y_train)

clf.score(X_test, y_test)
```
