```python
from nltk.tokenize import word_tokenize,sent_tokenize
from nltk.stem import PorterStemmer

f = open(r'D:\NLP\Austen_Emma.txt') 

text = f.read()

w = word_tokenize(text)

ps = PorterStemmer()

w_stemmed = []

for i in w:
    w_new = ps.stem(i)
    w_stemmed.append(w_new)

print(len(set(w)))
print(len(set(w_stemmed)))
    
```
