## 计算平均句长

```python
import jieba,re  #导入jieba分词库，导入正则库
path = r'D:/博士 第四学期/第五届中国语料库语言学大会 CLIC2020/生物工程学报2010-2019 摘要.txt' #文件的路径
```

```python
#定义分句
def sentence_cut(path):  
    with open(path,'r',encoding='ansi') as f:
        text = f.read()
    sentence = re.split('。|！|\!|\.|？|\?',text)
    sentences = [sent.strip() for sent in sentence]
    sentences_list = [list(jieba.cut(sent)) for sent in sentences]
    return sentences_list
all_sentences = []
sentences = sentence_cut(path)
all_sentences += sentences
print('文本句子的总数为：{}'.format(len(all_sentences)))
```

```python
def average_sentence():
    with open(path,'r',encoding='ansi') as f:
        text = f.read()
    a = len(all_sentences)  #计算文本中所有的句子
    b = len(text)  #计算文本的总字符
    result = b/a
    print('文本总字符数:{}'.format(len(text)))
    print('文本平均句长为:{}'.format(result))
    
average_sentence()
```

## 计算超纲词

```python
import pkuseg  #导入北大分词软件
import re #导入正则
seg = pkuseg.pkuseg() # 以默认配置加载模型
with open(r"D:/博士 第四学期/第五届中国语料库语言学大会 CLIC2020/世界汉语教学2010-2019.txt",encoding='ansi')as f:
    text = f.read() #读取文档赋值给变量text
```

```python
text = re.sub('[a-zA-Z\d\,\.;\'，。‘’*“”；、：！？《》\-（）()/\n\t]','',text) #去掉文章中的标点符号
custom_seg = pkuseg.pkuseg(user_dict='D:/NLP/project3/data/shijiehanyujiaoxue2010-2019keyword.txt')     #添加用户自定义词典
word1 = custom_seg.cut(text)
word_list = list(word1)
```

```python
from collections import Counter
word_freq = dict(Counter(word_list))
```

```python
list1= sorted(word_freq.items(),key=lambda x:x[0])  #x[0]表示降序
#print(list1)
```

```python
list1 = set(word1)
```

```python
import pandas as pd
data_1 = list1
data_2 = pd.read_excel('D:/NLP/project_1/data/新HSK一至六级词汇表(2011年).xls')
data_2.head()
```

```python
names = ['词汇']
for i in range(1,6):
    name = '词汇.%s'%i
    names.append(name)
names
```

```python
words_2 = []
for name in names:
    words = list(data_2[name].map(str))
    words_2 += words
words_2 = [word.strip() for word in words_2]
words_2 = set(words_2)
words_2.remove('nan')
len(words_2)
```

```python
words_1 = data_1
```

```python
words_set_1 = words_1 - words_2    # 只属于汉语教学的词语
words_set_2 = words_2 - words_1    # 只属于六级词汇的词语
common_words = words_1 & words_2   # 共有的词语
```

```python
print('list1的总单词数:{}'.format(len(list1))) 
print('words_set1的词语的个数：{}'.format(len(words_set_1)))
print('只属于六级词汇的词语的个数：{}'.format(len(words_set_2)))
print('共有的词语的个数：{}'.format(len(common_words)))
```
