```python
import pkuseg
import os
import re
path = '../data/original_text/legal' #所有文件的路径
folders = os.listdir(path)#获取所有的文件夹名称，返回名称列表
seg = pkuseg.pkuseg()
```

```python
folders#查看文件夹下面的文件
```

```python
#定义词语抽取
def word_extract(path):
    #打开txt文件
    with open(path,'r',encoding = 'gb18030',errors='ignore') as f:
        text=f.read()
    text = re.sub('[a-zA-Z\d\,\.;\'，①①③④⑤⑥⑦⑧⑨Ｒ≤Ⅴ↑＆∑Ｄ!"#$%&*+:。Ｅｘｌ‘’．“”；· 、：！Ｉ》:《＊ＰＯＡｐｒｏｄｕｃｔｉｏｎ－ｏｒｉｅｎｔｅｄａｐｐｒｏａｃｈ？\-（）()/\n\t　]','',text)  # 去掉文本中的字母、数字、标点、换行符
    words = list(seg.cut(text))
    return words    
```

```python
words_list = []
for folder in folders:
    files_path = path + '\\' + folder #某一文件夹路径
    files = os.listdir(files_path) #文件夹名称列表
    for f in files:
        f_path = files_path + '\\' + f #某一文件路径
        words = word_extract(f_path) #抽取词语
        words_list += words #将所有的词语放人到列表中
    print('{} have been extract'.format(folder))
print('词语总数为:{}'.format(len(words_list)))
word_set = set(words_list) #去掉重复的词语之后的词典
print('词典大小为:{}'.format(len(word_set)))
#words_list[:5000] #查看前5000
```

**生成词云图**

```python
from matplotlib import pyplot as plt
from wordcloud import WordCloud
from PIL import Image
import numpy as np

font = r'C:\Windows\Fonts\simhei.TTF'
graph_path = r'D:\NLP\project_1\词云图'

string = ' '.join(final_word)
img = Image.open(graph_path+r'\背景图.jpg')
img_array = np.array(img) #将图片转换为数组
stopword=['一个','不同','第二','可以','不是','例如','没有'] #设置停用词
wc = WordCloud(
    background_color='white',
    width=1600,
    height=1200,
    mask=img_array,
    font_path=font,
    stopwords=stopword,
    collocations=False
)
wc.generate_from_text(string)
plt.imshow(wc)
plt.axis('off')
plt.show()
wc.to_file(graph_path+r'词云图_2.png')
```
