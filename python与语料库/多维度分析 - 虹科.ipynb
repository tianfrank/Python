```python
import pkuseg
import re

with open('../data/student_all_together.txt','r',encoding='utf-8') as f:
    text = f.read()
    text = re.sub('[a-zA-Z\d\,\;\'，○——？！……①②③‘’“”；、：\-（）()/\n\t]',' ',text)  # 去掉文本中的字母、数字、标点、换行符，不缩进表示在文件外操作
seg = pkuseg.pkuseg(postag=True)           # 以默认配置加载模型,post=True表示的是标注词性
text = seg.cut(text)  # 进行分词[(),(词，v)]
#print(text)  #返回的数据类型是列表，里面是元组

#相对路径是以代码所在的路径为基准，绝对路径是在硬盘所在的位置
```

```python
#text = [('苹果', 'n'), ('是','v'),('桌子', 'n'), ('打','v'),('好', 'adj'), ('吃','v')]，这是pkuseg分词后返回的数据类型，列表里面套元组
# pos_dict = {'v':['打','吃'], 'n': ['名字']}

pos_dict = {}  #建立一个装词性的字典
total_words = len(text) #统计文本的所有单词【里面要去掉标点符号】
for item in text:   #因为结巴分词后返回的text格式为元组
    pos = item[1] #所以元组里面1号位置为词性
    word = item[0] #元组里面0号位置为单词
    if pos not in pos_dict: #如果词性标签不在pos_dict字典里面
        pos_dict[pos] = [word] #就新建一个列表
    else:
        pos_dict[pos].append(word)#如果词性标签在post字典里面，就一个一个添加到pos_dict
#pos_dict  #查看掉用pos_dict字典
```

```python
#每类词出现的频次
pos_count = {}   #{"V":20,"N":30}
for pos,word_list in pos_dict.items():
    pos_count[pos]=len(word_list)
pos_count=sorted(pos_count.items(),key=lambda x: x[1],reverse=True) #根据字典值排序
pos_count  #sorted排序后返回的是元组的列表，如果转成字典要用dict
```

```python
#对动词按照频率做一个排序
from collections import Counter

#统计词频
#{"v":{打，20}，{是，10}}
word_pos_freq = {}  #新建
for pos, word_list in pos_dict.items():
    freq_dict = {}
    for word, freq in dict(Counter(word_list).most_common()).items():
        freq_dict[word] = freq 
    word_pos_freq[pos] = freq_dict
word_pos_freq
```

```python
#批量保存
for pos,word_freq in word_pos_freq.items():
    with open('../result/{}.txt'.format(pos),'w',encoding='utf-8') as f:  #..返回上一级的路径，result往后走路径
        for word,freq in word_freq.items():
            f.write(word+','+str(freq)+'\n')          
```

```python
#各类词占总词的比例
pos_count = {} #定义一个词频统计的字典
for key, value in pos_dict.items(): #key为pos_count的键，就是词性标签；value为pos_count的值，value里面是相同词性的词的集合
    pos_count[key] = len(value) / total_words #len(value)表示的是统计相同词的词集合的总数，total_words是统计文本所有单词
pos_count

#保存结果
with open('pos_count.txt','w',encoding='utf-8') as f:
    for key,value in pos_count.items():
        f.write(key+','+str(value)+'\n')
```

```python
# {'v':{'我':50%,}}
from collections import Counter
word_pos_percent = {}  #新建
for pos, word_list in pos_dict.items():
    percent_dict = {}
    for word, freq in dict(Counter(word_list).most_common()).items():
        percent_dict[word] = freq / len(word_list)
    word_pos_percent[pos] = percent_dict
word_pos_percent
```

```python
#保存结果
with open('word_pos_percent','w',encoding='utf-8') as f:
    for word,freq in word_pos_percent.items():
        f.write(word+','+str(freq)+'\n')
```
