```python
with open(r'../data/korean_ma/Lee.txt','r',encoding='utf-8') as f:
    content = f.read()
    
import pandas as pd
import numpy as np
import os
import re
content=re.sub("\t山东大学硕士学位论文\n",'',content)

def seperation_txt(text):
    abstract = re.findall("摘要.*关键词",content,flags=re.S)
    abstract = "".join(abstract)  #把列表里面的内容转为字符串
    #第一章
    introduction = re.findall("第一章.*第二章",content,flags=re.S)  #抽出摘要,findall返回的数据类型为列表
    introduction ="".join(introduction)
    #第二章到结语
    body = re.findall("第二章.*结语|第二部分.*结语",content,flags=re.S) #|表示完整的正则符号
    body ="".join(body)
    #第三部分结论到参考文献
    conclusion = re.findall("结语\n.*?附录|结论\n.*?附录",content,flags=re.S) #。?表示非贪婪模式匹配
    conclusion ="".join(conclusion)
    #第四部分 参考文献到致谢
    acknowledgements = re.findall("致谢\n.*",content,flags=re.S)  #加问号是非贪婪模式
    acknowledgements ="".join(acknowledgements)    
    return abstract,introduction,body,conclusion,acknowledgements  #返回的是列表,列表的内容转字符串用.join()

#读取路径的文件，调用函数，将得到抽取的五个结果放入字典里
path = r'D:\NLP\project5\study\data\korean_ma\korea_ma_remove'
files = os.listdir(path)
final_results={"title":[],"abstract":[],"introduction":[],"body":[],"conclusion":[],"acknowledgements":[]}
for file in files:
    file_path = path+'\\'+file
    with open(file_path,'r',encoding='utf-8',errors = 'ignore') as f:
        text = f.read()  #一个一个打开
    abstract,introduction,body,conclusion,acknowledgements=seperation_txt(text) #一个函数有多个返回值用逗号隔开
    final_results['title'].append(file)
    final_results['abstract'].append(abstract)
    final_results['introduction'].append(introduction)
    final_results['body'].append(body)
    final_results['conclusion'].append(conclusion)
    final_results['acknowledgements'].append(acknowledgements)
    
#final_results
 
```

```python
#字典转dataframe
df = pd.DataFrame(final_results)
df.to_csv('final_results.csv',index=False)  #index是索引
```
