```python
**python 输入一行字符，分别统计出其中英文字母、空格、数字和其它字符的个数。**
#https://www.cnblogs.com/python-xkj/p/9225985.html
```

```python
#方法1
char = input("请输入字符:")
letters = []
space = []
number = []
others = []
for i in iter(char):
    if i.isalpha():
        letters.append(i)
    elif i.isspace():
        space.append(i)
    elif i.isdigit():
        number.append(i)
    else:
        others.append(i)
print('''
字母:{},个数:{}
空格:{},个数:{}
数字:{},个数:{}
其他:{},个数:{}'''\
      .format(letters,len(letters),space,len(space),number,len(number),others,len(others)))
        
```

```python
#方法2
char = input("请输入字符:")
letters = 0
space = 0
number = 0
others = 0
for i in char:
    if i.isalpha():
        letters +=1
    elif i.isspace():
        space +=1
    elif i.isdigit():
        number +=1
    else:
        others +=1
print('字母=%d'%letters)
print('空格=%d'%space)
print('数字=%d'%number)
print('其他=%d'%others)      
```
