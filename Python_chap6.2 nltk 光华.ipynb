```python
from nltk.corpus import gutenberg
for fileid in gutenberg.fileids():
    num_chars = len(gutenberg.raw(fileid))
    num_words = len(gutenberg.words(fileid))#表示有多少个词
    num_sents = len(gutenberg.sents(fileid)) #sents表示有多少个sentsence
    num_vocab = len(set(w.lower() for w in gutenberg.words(fileid)))#.lower变换成小写
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocab), fileid)
    
    #结果表示一个单词的长度有平均5个字母，一个句子有25个单词，平均用了26次
```

## Gender Identification

**1. Look at the data**

```python
from nltk.corpus import names
labeled_names = ([(name, 'male') for name in names.words('male.txt')] +
[(name, 'female') for name in names.words('female.txt')])
import random
random.shuffle(labeled_names) #打乱
```

```python
type(labeled_names) #查看类型
```

```python
labeled_names[:10] #查看
```

**2. 提取特征**

```python
def gender_features(word):
    return {'last_letter': word[-1]} #字典，：前为键，后为值
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labeled_names]
```

```python
featuresets[:5]
```

```python
len(featuresets)
```

**3. 建立分类器**

```python
import nltk
train_set, test_set = featuresets[500:], featuresets[:500] #分了2类，前500为训练，后500为测试
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

**4. 预测与评估**

```python
classifier.classify(gender_features('Jim'))
```

```python
classifier.classify(gender_features('Jessica'))
```

```python
nltk.classify.accuracy(classifier, test_set) #查看准确率
```

```python
classifier.show_most_informative_features(10) #查看最有用的信息
```

**Error Analysis**

```python
train_names = labeled_names[1500:]
devtest_names = labeled_names[500:1500]
test_names = labeled_names[:500]
```

```python
train_set = [(gender_features(n), gender) for (n, gender) in train_names]
devtest_set = [(gender_features(n), gender) for (n, gender) in devtest_names]
test_set = [(gender_features(n), gender) for (n, gender) in test_names]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, devtest_set))
```

```python
errors = []
for (name, tag) in devtest_names:
    guess = classifier.classify(gender_features(name))
    if guess != tag:
        errors.append( (tag, guess, name) )
```

```python
errors
```

```python
def gender_features(word):
     return {'suffix1': word[-1:],
             'suffix2': word[-2:]} #增加特征，选择最后1个和最后2个字母
```

```python
train_set = [(gender_features(n), gender) for (n, gender) in train_names]
devtest_set = [(gender_features(n), gender) for (n, gender) in devtest_names]
classifier = nltk.NaiveBayesClassifier.train(train_set)
print(nltk.classify.accuracy(classifier, devtest_set))
```

## Document Classification

```python
from nltk.corpus import movie_reviews
documents = [(list(movie_reviews.words(fileid)), category)
              for category in movie_reviews.categories()
              for fileid in movie_reviews.fileids(category)]
random.shuffle(documents)
```

```python
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())
word_features = list(all_words)[:2000]

def document_features(document):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features
```

```python
print(document_features(movie_reviews.words('pos/cv957_8737.txt'))) 
```

```python
featuresets = [(document_features(d), c) for (d,c) in documents]
train_set, test_set = featuresets[100:], featuresets[:100]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

```python
classifier.show_most_informative_features(5)
```
