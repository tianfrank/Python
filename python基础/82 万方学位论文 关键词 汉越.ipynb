```python
import os
import re

import requests
import xlrd
import xlutils.copy
import xlwt
from bs4 import BeautifulSoup
from requests import RequestException


def get_html(url):
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/65.0.3325.162 Safari/537.36'
        }
        response = requests.get(url, headers=headers)
        if response.status_code == 200:
            return response.text
        return None
    except RequestException as e:
        print(e)
        return None


def parse_html(html, url):
    soup = BeautifulSoup(html, 'lxml')
    # 题目
    # title = soup.find(name='div', attrs={'class': 'title'}).text
    title = soup.find(name='font', attrs={'style': 'font-weight:bold;'}).text
    print(title)
    # 摘要
    abstract = soup.find(name='div', attrs={'class': 'abstract'}).textarea
    if abstract:
        abstract = abstract.text
    print(abstract)
    # 关键词
    keyword = soup.find_all(name='a', attrs={'title': '知识脉络分析', 'href': '#'})
    keywords = ''
    for word in keyword:
        keywords = keywords + word.text + ';'
    print(keywords)
    # 作者
    author = soup.find(name='a', id='card01').text
    print(author)
    # 学位授予单位
    unit = soup.find(name='a', attrs={'title': '知识脉络分析', 'href': 'javascript:void(0)'}).text
    print(unit)
    # 学位
    pattern = re.compile('授予学位.*?<div class="info_right author">(.*?)</div>', re.S)
    degree = re.findall(pattern, html)[0]
    print(degree)
    # 导师
    teacher = soup.find(name='a', id='card00').text
    print(teacher)
    # 学位年度
    pattern = re.compile('学位年度.*?<div class="info_right author">(.*?)</div>', re.S)
    date = re.findall(pattern, html)[0]
    print(date)
    paper = [title, abstract, keywords, author, unit, degree, teacher, date, url]
    print(paper)
    return paper


def save_p(paper):
    if not os.path.exists('学位论文.xls'):
        wb = xlwt.Workbook()
        sheet = wb.add_sheet('sheet1')
        title = ['题目', '摘要', '关键词', '作者', '学位授予单位', '学位', '导师', '学位年度', '链接']
        for i in range(len(title)):
            sheet.write(0, i, title[i])
        wb.save('学位论文.xls')
    wb = xlrd.open_workbook('学位论文.xls')
    sheet = wb.sheet_by_index(0)
    rows = sheet.nrows
    print(rows)
    ws = xlutils.copy.copy(wb)
    sheet = ws.get_sheet(0)
    for i in range(len(paper)):
        sheet.write(rows, i, paper[i])
    ws.save('学位论文.xls')


def main(url):
    html = get_html(url)
    paper = parse_html(html, url)
    save_p(paper)

url_list = []

    
# if __name__ == '__main__':    
#     url = [i for i in url_list] #师傅说在jupyter里面不用写if __name=='__main__':

for url in url_list:
    main(url)
```
