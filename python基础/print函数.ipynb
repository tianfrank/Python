#print() 函数的用法有以下几种：单刀赴会—不带引号，黄袍加身—搭配单引号、双引号、三引号，让我们一睹风采

```python
print(1234) #默认是数字，所有不用加引号
print('哪吒') #字符要加引号告诉计算机这是引号
print("哪吒") #双引号与单引号是一样的效果
print('''哪吒，
托塔李天王家的三太子，
变化三头六臂又或三头八臂，
百邪不侵专克摄魂夺魄的莲花化身''') #三引号可以实现打印多行

#实现换行的另一种办法，使用换行符\n
print('哪吒，\n托塔李天王家的三太子，\n变化三头六臂又或三头八臂，\n百邪不侵专克摄魂夺魄的莲花化身')
```
