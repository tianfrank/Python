```python
#函数作为返回值，是高阶函数
#在函数内部返回一个函数，返回的这个函数就是返回函数
#函数嵌套
def show():
    def show1():
        print("天地玄黄，宇宙洪荒")
    return show1
#newfunc就是show1
new_func = show()
new_func()
#返回函数也是高阶函数的一种
```
