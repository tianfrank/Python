# 1向量

```python
#1创建向量
import numpy as np
np.array((2,4))
```

```python
#2向量的计算
# v = [2,4]
# u = [3,1]
# 2+3 = 5
# 4+1 = 5
#向量的加法
V=np.array((2,4))
U=np.array((3,1))
V-U
```

##向量乘以常量

```python
V = np.array((2,4))
V*1/3
```

##向量的点乘（内积）

```python
V=np.array((2,4))
U=np.array((3,1))

#2*3+4*1=10
np.dot(V,U) #np.dot表示点乘，对应位置相乘之和
```

```python
#求矩阵A[2,5,4]与矩阵B[3,4,5]的内积？

import numpy as np
A = np.array((2,5,4))
B = np.array((3,4,5))
A*B
```

##向量的叉乘（外积、向量积）

```python
# V = [a1,a2,a3]
# U = [b1,b2,b3]

# P = [a2*b3-a3*b2,a3*b1-a1*b3,a1*b2-a2*b1]
V = np.array((1,2,3))
U = np.array((4,5,6))
#2*6 - 3*5 = -3
#3*4-1*6=6
#2*4-1*5=-3

np.cross(V,U) #np.cross表示叉乘
```

##向量的哈达玛积

```python
V=np.array((1,2,3))
U=np.array((4,5,6))
V*U
```

### 向量小结

```python
V*U #哈达玛积
np.dot()#内积
np.cross()#外积
```

# 2矩阵

### 2.1什么是矩阵

**在数学中：矩阵是一个按照长方阵排列的复数或实数集**

**线性代数中：矩阵是描述线性代数中线性关系的参数，即矩阵是一个线性变换，可以将一些向量转换为另一些向量。**



```python
import numpy as np

np.mat(np.random.randint(1,10,size=(3,4))) #mat=matrix,randint指随机整数，1-10的范围，3行4列的矩阵
```

### 2.2矩阵的计算方法

#### 2.2.1矩阵的加减法

```python
A = np.mat(np.random.randint(1,10,size=(3,4)))
B = np.mat(np.random.randint(1,10,size=(3,4)))

print(A)
print('--------------')
print(B)

A - B
```

####  2.2.2矩阵与数的乘法

```python
A = np.mat(np.random.randint(1,10,size=(3,4)))
print(A)
A*3
```

#### 2.2.3矩阵与向量相乘

```python
A = np.mat([[1,2,3],[4,5,6],[7,8,9]])

V = np.array([[2],[2],[2]])

print(A)
print(V)
A*V
```

#### 2.2.4矩阵与矩阵相乘

```python
A = np.mat([[1,2],[3,4]])
B = np.mat([[1,2],[3,4]])
print(A)
print(B)
A*B
```

```python
np.multiply(A,B)
```
