```python
#装饰器：一些操作，文件
#装饰器：本质上也是一个函数，可以给原函数的功能进行扩展
#不改变原函数额定义和调用操作

def display():
    print("静待春暖花开")
#在AAA前面加上————————（装饰器本质是一个闭包）
#display
#装饰器函数
def decorator(new_func):
    #定义
    def inner():
        print("_______")
        #调用传进来的函数
        new_func()
    return inner
display = decorator(display)
display()
```
