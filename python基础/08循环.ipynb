```python
#输入100条hello world
#死循环，一直满足条件就死循环
i = 100
while i <= 100:
    print("hello word")
i += 1
```

```python
#打印0-5,执行5次循环

num = 0  #赋值初始值
while num < 5:
    print(num)
    num += 1 # 相当于num = num + 1 等于 num +=1
print("-----------------------------")
```

```python
#for循环，结合range来使用（range代表一个范围）
#range(0-4)
#默认:起始数据默认为0，结束必写，步长默认1
for value in range(5):
    print(value)
```

```python
#(0,6,1) 0表示起始数据，6代表结束数据（结束数据不执行循环）。1为步长（跳跃的范围）
for value in range(0,6,3):
    print(value)
```

```python
#for循环和while循环可以结合else语句来使用
print("_________________")
num = 5
while num>=1:
    print(num)
    num -= 1
else:
    print("循环结束")
```
