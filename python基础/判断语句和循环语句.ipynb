```python
age = int(input("请输入你的年龄"))
print(age)
#默认输入的文字，返回值的类型是str类型
print(type(age))

#if判断语句：满足条件才会执行代码
#只有同类型才可以比较
#and两面条件都满足，才去执行
#or满足其中一个条件就去执行

if age>= 15 and age <= 17:  
    print("青少年")
elif age>= 17 and age <= 40:
    print("中年")
elif age <=14:
    print("小朋友")
else:
    print("老年")
   
```
