# 定义变量

```python
#变量通俗就是存储数据的容器
a = 10  #把10赋值给变量a,a=10
b = 20  #把20赋值给变量b，b=20
result = a + b
print(result)
```

# 数据类型的转换

```python
#1.字符串--->整型
a = '123'
int(a)
print(type(a))

"""int()函数将字符串转换为整数类型，需要注意的是字符串需要兼容数字
a = '说的说的'
int(a) #这样就属于不兼容
#type()函数可以帮助我们查看当前数据的类型
print(type(a))

"""
```

```python
#2.整型-->字符串
a = 123
str(a)
print(type(a))
```

```python
#3.字符串转换为浮点型
a = '1.23'
float(a)
print(type(a))
```

```python
小结：数据类型的转换函数，int(),str(),float()
```
